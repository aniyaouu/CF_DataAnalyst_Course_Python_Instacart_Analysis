
## 4.3 Line Chart for Expenditure


```python
# Check whether thereâ€™s a difference in expenditure (the â€œpricesâ€ column) depending on the hour of the day. 
# (Hint: To check this, you need to use an accurate sample for your line chart!)
```

```python
# Creating a sample
np.random.seed(4)
dev = np.random.rand(len(ords_prods_deps_cust)) <= 0.7

# Storing 70% of the sample in the data frame big
big = ords_prods_deps_cust[dev]

# Storing 30% of the sample in the data frame small
small = ords_prods_deps_cust[~dev]
```

```python
len(ords_prods_deps_cust)
```

```python
len(big) + len(small)
```

```python
# Reducing data frame to necessary columns only
df_3 = small[['order_hour','prices']]
```

```python
df_3
```

```python
# Grouping data to calculate average expenditure per hour
hourly_prices = df_3.groupby('order_hour')['prices'].mean()

# Creating the line chart
plt.figure(figsize=(10, 6))
plt.plot(hourly_prices.index, hourly_prices.values, marker='o', color='blue', linestyle='-')

# Adding titles and labels
plt.title('Average Expenditure by Hour of the Day', fontsize=16)
plt.xlabel('Hour of the Day', fontsize=12)
plt.ylabel('Average Prices', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(range(0, 24), fontsize=10)  # Ensuring all 24 hours are displayed on x-axis

# Save the chart
plt.savefig(os.path.join('/Users/xxx/Documents/Instacart Basket Analysis - 2025-01-05/', '04 - Analysis','Visualizations', 'line_order_hour_prices_formatted.png'))

# Display the chart
plt.show()
```

```python
# Creating the line chart
line_oh_p = sns.lineplot(data = df_3, x = 'order_hour',y = 'prices')
```

```python
# Saving the basic line chart
line_oh_p.figure.savefig(os.path.join('/Users/xxx/Documents/Instacart Basket Analysis - 2025-01-05/', '04 - Analysis','Visualizations', 'line_order_hour_prices_basic.png'))
```


## 4.4 Exploratory analysis of customer demographics


```python
# Determine connection between age and family situation 
# Creating a line chart exploring the connections between age and number of dependents
```

```python
# Creating a sample
np.random.seed(4)
dev = np.random.rand(len(ords_prods_deps_cust)) <= 0.7

# Storing 70% of the sample in the data frame big
big = ords_prods_deps_cust[dev]

# Storing 30% of the sample in the data frame small
small = ords_prods_deps_cust[~dev]
```

```python
# Reducing data frame to necessary columns only
df_4 = small[['age','num_dependants']]
```

```python
df_4
```

```python
# Grouping data to calculate average expenditure per hour
avg_num_dependants_for_age = df_4.groupby('age')['num_dependants'].mean()

# Creating the line chart
plt.figure(figsize=(10, 6))
plt.plot(avg_num_dependants_for_age.index, avg_num_dependants_for_age.values, marker='o', color='blue', linestyle='-')

# Adding titles and labels
plt.title('Average Number of Dependants by Age', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Number of Dependants', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
#plt.xticks(range(0, 90), fontsize=10) 

# Save the chart
plt.savefig(os.path.join('/Users/xxx/Documents/Instacart Basket Analysis - 2025-01-05/', '04 - Analysis','Visualizations', 'line_age_num_dependants_formatted.png'))

# Display the chart
plt.show()
```

```python
# Creating a basic line chart
line_age_dependants = sns.lineplot(data = df_4, x = 'age',y = 'num_dependants')
```
