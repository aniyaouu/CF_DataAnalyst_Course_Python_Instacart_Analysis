## 3.2 Creating "busiest_day" column

```python
# Display data frame

ords_prods_deps_cleaned
```

```python
# Count the number of orders for each weekday
weekday_counts = ords_prods_deps_cleaned['order_weekday'].value_counts()

# Identify the busiest and least busy days
busiest_day = weekday_counts.idxmax()  # The day with the highest orders
least_busy_day = weekday_counts.idxmin()  # The day with the lowest orders

# Assign categories
ords_prods_deps_cleaned['busiest_day'] = ords_prods_deps_cleaned['order_weekday'].apply(
    lambda x: 'Busiest day' if x == busiest_day else ('Least busy day' if x == least_busy_day else 'Regular day')
)

#Frequency in %
price_range_percentages = ords_prods_deps_cleaned['busiest_day'].value_counts(normalize=True) * 100
print(price_range_percentages)

# Display the first few rows to check the new column
print(ords_prods_deps_cleaned[['order_weekday', 'busiest_day']].head())
```

# 4. Identify the busiest hours of the day

```python
# Rather than by hour label periods of time: “Most orders,” “Average orders,” and “Fewest orders.” 
# Create a new column containing these labels called “busiest_period_of_day.”
```

```python
# Count the number of orders for each hour
hourly_counts = ords_prods_deps_cleaned['order_hour'].value_counts().sort_index()

# Calculate quartiles
q1 = hourly_counts.quantile(0.25)  # Bottom 25% threshold
q3 = hourly_counts.quantile(0.75)  # Upper 25% threshold

# Assign categories based on quartiles
ords_prods_deps_cleaned['busiest_period_of_day'] = ords_prods_deps_cleaned['order_hour'].apply(
    lambda x: 'Most orders' if hourly_counts[x] >= q3 else 
              ('Fewest orders' if hourly_counts[x] <= q1 else 'Average orders')
)

#Frequency in %
price_range_percentages = ords_prods_deps_cleaned['busiest_period_of_day'].value_counts(normalize=True) * 100
print(price_range_percentages)

# Display first few rows to check the new column
print(ords_prods_deps_cleaned[['order_hour', 'busiest_period_of_day']].head())
```
