
# 4. Creating a loyalty flag for existing customers using the transform() and loc() functions


```python
# If the maximum orders the user has made is over 40, then the customer will be labeled a “Loyal customer.”
# If the maximum orders the user has made is over 10 but less than or equal to 40, then the customer will be labeled a “Regular customer.”
# If the maximum orders the user has made is less than or equal to 10, then the customer will be labeled a “New customer.”
```

```python
# Create “max_order” column
ords_prods_deps['max_order'] = ords_prods_deps.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Create “max_order” column
ords_prods_deps['max_order'] = ords_prods_deps.groupby(['user_id'])['order_number'].transform('max')

#Frequency in %
price_range_percentages = ords_prods_deps['max_order'].value_counts(normalize=True) * 100
print(price_range_percentages)
```

```python
ords_prods_deps.head(100)
```


# 4.1 Deriving Columns with loc()


```python
# If the maximum orders the user has made is over 40, then the customer will be labeled a “Loyal customer.”
ords_prods_deps.loc[ords_prods_deps['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'

# If the maximum orders the user has made is over 10 but less than or equal to 40, then the customer will be labeled a “Regular customer.”
ords_prods_deps.loc[(ords_prods_deps['max_order'] <= 40) & (ords_prods_deps['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'

# If the maximum orders the user has made is less than or equal to 10, then the customer will be labeled a “New customer.”
ords_prods_deps.loc[ords_prods_deps['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```


# 6. Create a spending flag for each user based on the average price across all their orders



#### If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”



#### If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”


```python
# Calculate the mean price of products purchased by each user
user_avg_prices = ords_prods_deps.groupby('user_id')['prices'].mean()

# Map the spending flag based on the criteria
spending_flag = user_avg_prices.apply(lambda x: 'Low spender' if x < 10 else 'High spender')

# Add the spending flag to the original data frame by merging
ords_prods_deps = ords_prods_deps.merge(spending_flag.rename('spending_flag'), on='user_id', how='left')

# Display the first few rows to verify
print(ords_prods_deps[['user_id', 'prices', 'spending_flag']].head())
```

```python
# Display unique values in the 'spending_flag' column
unique_values = ords_prods_deps['spending_flag'].unique()
print(unique_values)
```


# 7. Determine frequent versus non-frequent customers


```python
# 1. If the median of “days_since_last_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
# 2. If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
# 3. If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”
```

```python
# Calculate the median of "days_since_last_order" for each user
user_median_days = ords_prods_deps.groupby('user_id')['days_since_last_order'].median()

# Create the order frequency flag based on the specified criteria
frequency_flag = user_median_days.apply(
    lambda x: 'Non-frequent customer' if x > 20 
              else 'Regular customer' if 10 < x <= 20 
              else 'Frequent customer'
)

# Add the frequency flag to the original data frame by merging
ords_prods_deps = ords_prods_deps.merge(frequency_flag.rename('order_frequency_flag'), on='user_id', how='left')

# Display the first few rows to verify
print(ords_prods_deps[['user_id', 'days_since_last_order', 'order_frequency_flag']].head())
```

```python
# Display unique values in the 'order_frequency_flag' column
unique_values = ords_prods_deps['order_frequency_flag'].unique()
print(unique_values)
```
